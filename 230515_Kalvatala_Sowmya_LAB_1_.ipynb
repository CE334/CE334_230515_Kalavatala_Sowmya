```python
import numpy as np
```

```python
# WGS84 parameters
a=6378137.0 #semi-major axis(in m)
f=1/298.257223563 #flattening factor
b=a*(1-f) #semi-minor axis(in m)
e2=2*f-f**2 #eccentricity squared




# function to convert deg to radians:
def convert(deg,min,sec):
  return np.deg2rad(deg+min/60+sec/3600)

# function to convert radians to deg min sec:
def rad_to_dms(rad,is_latitude=True):
    dd=np.rad2deg(rad)
    if is_latitude:
        suffix="N" if dd>=0 else "S"
    else:
        suffix="E" if dd>=0 else "W"
    d=int(abs(dd))
    m=int((abs(dd)-d)*60)
    s=(abs(dd)-d-m/60)*3600
    return f"{d}° {m}' {s:.0f}\""
# Fixed: Convert DMS to radians using the 'convert' function
lat_k = convert(26, 26, 59.7228)
lon_k = convert(80, 19, 54.7356)
r_k= 126.630

lat_n = convert(15,29,19) # Placeholder for lat_n
lon_n = convert(78,29,11) # Placeholder for lon_n
r_n = 216   # Placeholder for r_n
```

```python
#QUESTION-1#
# Geodetic to Cartesian
def geod_to_cart(lat, lon, h):
    N = a / np.sqrt(1 - e2 * np.sin(lat)**2)

    x = (N + h) * np.cos(lat) * np.cos(lon)
    y = (N + h) * np.cos(lat) * np.sin(lon)
    z = (N * (1 - e2) + h) * np.sin(lat)

    return x, y, z
xk,yk,zk = geod_to_cart(lat_k,lon_k,r_k)
xn,yn,zn = geod_to_cart(lat_n,lon_n,r_n)
print("geodetic to cartesian of kanpur: \n")
print("X:",xk)
print("Y:",yk)
print("Z:",zk)
print("geodetic to cartesian of nandyal: \n")
print("X:",xn)
print("Y:",yn)
print("Z:",zn)

# Cartesian to Geodetic (iterative)
def cart_to_geod(x, y, z):
    lon = np.arctan2(y, x)
    p = np.sqrt(x**2 + y**2)

    lat = np.arctan2(z, p * (1 - e2))   # initial value

    for _ in range(5):
        N = a / np.sqrt(1 - e2 * np.sin(lat)**2)
        h = p / np.cos(lat) - N
        lat = np.arctan2(z, p * (1 - e2 * N / (N + h)))

    return lat, lon, h
lat_sk,lon_sk,r_sk=cart_to_geod(xk,yk,zk)
lat_sn,lon_sn,r_sn=cart_to_geod(xn,yn,zn)
print("cartesian to geodetic of kanpur: \n")
print("LAT:",rad_to_dms(lat_sk))
print("LON:",rad_to_dms(lon_sk))
print("R:",r_sk)
print("cartesian to geodetic of nandyal: \n")
print("LAT:",rad_to_dms(lat_sn))
print("LON:",rad_to_dms(lon_sn))
print("R:",r_sn)
```

```python
# QUESTION-2: #

# function to transform spherical coordinates into cartesian coordinates:
def sph_to_cart(r,lat,lon):
    # r:radial distance from the origin, lat:geocentric latitude, lon: geocentric longitude
    x=r*np.cos(lat)*np.cos(lon) #projection to x
    y=r*np.cos(lat)*np.sin(lon) #projection to y
    z=r*np.sin(lat) #projection to z(Earth's rotation axis)
    return x,y,z
xk,yk,zk = sph_to_cart(r_k,lat_k,lon_k)
xn,yn,zn = sph_to_cart(r_n,lat_n,lon_n)
# function to transform cartesian coordinates into spherical coordinates:
def cart_to_sph(x,y,z):
    # x,y,z are cartesian coordinates
    r=np.sqrt(x**2+y**2+z**2)
    lat=np.arcsin(z/r)
    lon=np.arctan2(y,x)
    return r,lat,lon
r_sk,lat_sk,lon_sk=cart_to_sph(xk,yk,zk)
r_sn,lat_sn,lon_sn=cart_to_sph(xn,yn,zn)
print("spherical to cartesian of kanpur: \n")
print("X:",xk)
print("Y:",yk)
print("Z:",zk)
print("spherical to cartesian of nandyal: \n")
print("X:",xn)
print("Y:",yn)
print("Z:",zn)
print("\n")
print("cartesian to spherical of kanpur: \n")
print("R:",r_sk)
print("LAT:",rad_to_dms(lat_sk))
print("LON:",rad_to_dms(lon_sk))
print("cartesian to spherical of nandyal: \n")
print("R:",r_sn)
print("LAT:",rad_to_dms(lat_sn))
print("LON:",rad_to_dms(lon_sn))
```

```python
# QUESTION-3: #

# we transform geodetic coordinates to Cartesian (x,y,z) and cartesian coordinates to Spherical (r,φ,λ)
def geod_to_sph(lat_g,lon_g,h,a_ell,r_sph=None):
    if r_sph is None: # Using the default radius if its not provided
        r_sph=a_ell
    # If the geodetic coordinates are given in DMS(deg,min,sec) or decimal degrees, they must be converted to radians first (using convert() or np.deg2rad) before calling this function.
    x,y,z=geod_to_cart(lat_g,lon_g,h) # x,y,z are the cartesian coordinates
    r_s,lat_s,lon_s=cart_to_sph(x,y,z) # r_s: radial distance, lat_s: geocentric latitude, lon_s: geocentric longitude
    return r_s,lat_s,lon_s
r_gk,lat_gk,lon_gk = geod_to_sph(lat_sk,lon_sk,r_sk,a)
r_gn,lat_gn,lon_gn = geod_to_sph(lat_sn,lon_sn,r_sn,a)
print("geodetic to spherical of kanpur: \n")
print("R:", r_gk)
print("LAT:",rad_to_dms(lat_gk))
print("LON:",rad_to_dms(lon_gk))
print("geodetic to spherical of nandyal: \n")
print("R:", r_gn)
print("LAT:",rad_to_dms(lat_gn))
print("LON:",rad_to_dms(lon_gn))
```

```python
# QUESTION-4: #

# function to transform geodetic coordinates into topocentric coordinates:
def geod_to_topo(lat_obs,lon_obs,h_obs,lat_tgt,lon_tgt,h_tgt):
    x_obs,y_obs,z_obs=geod_to_cart(lat_obs,lon_obs,h_obs)
    x_tgt,y_tgt,z_tgt=geod_to_cart(lat_tgt,lon_tgt,h_tgt)
    # Computing the displacement vector (dx,dy,dz)
    dx=x_tgt-x_obs
    dy=y_tgt-y_obs
    dz=z_tgt-z_obs
    # Applying the rotation matrix to get East-North-Up(ENU)
    de=-np.sin(lon_obs)*dx+np.cos(lon_obs)*dy
    dn=-np.cos(lon_obs)*np.sin(lat_obs)*dx-np.sin(lon_obs)*np.sin(lat_obs)*dy+np.cos(lat_obs)*dz
    du=np.cos(lon_obs)*np.cos(lat_obs)*dx+np.sin(lon_obs)*np.cos(lat_obs)*dy+np.sin(lat_obs)*dz
    return de,dn,du
# Defining Geodetic coordinates for Nuzvid(Topocentre)
lat_n_rad=convert(15,29,19)
lon_n_rad=convert(78,29,11)
h_n=100.000
# Defining Geodetic coordinates for Kanpur
lat_cnb_rad=convert(26,26,59.7228)
lon_cnb_rad=convert(80,19,54.7356)
h_cnb=126.630
h_sat=h_cnb+800000 #Since satellite is flying exactly overhead Kanpur at an altitude of 800 km
e_sat,n_sat,u_sat=geod_to_topo(lat_n_rad,lon_n_rad,h_n,lat_cnb_rad,lon_cnb_rad,h_sat)   # Computing topocentric coordinates

print("Topocentric coordinates of the satellite(relative to Nuzvid):")
print(f"East:{e_sat:.3f}m")
print(f"North:{n_sat:.3f}m")
print(f"Up:{u_sat:.3f}m")
```
