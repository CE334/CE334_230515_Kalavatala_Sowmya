```python
#2nd question
import numpy as np

# Creating a random 5x5 non-singular matrix
# We will try to create a random matrix until it's non-singular
# (i.e., its determinant is not zero).
# For large random matrices, the probability of being singular is very low.
a = np.random.rand(5, 5)
while np.linalg.det(a) == 0:
    a = np.random.rand(5, 5)

print("Matrix a (non-singular):\n", a)
```

```python
#5th question
# Copying matrix a into b
b = a.copy()

# If element is greater than 2, multiply it by 10
for i in range(a.shape[0]):
    for j in range(a.shape[1]):
        if a[i, j] > 2:
            b[i, j] = a[i, j] * 10

print("Matrix b after condition:\n", b)
```

```python
#13th question
def derivative(f, a, method='central', h=0.01):
    """
    This function computes the numerical derivative of f(x)
    at point a using forward, backward or central difference.
    """
    if method == 'forward':
        return (f(a + h) - f(a)) / h
    elif method == 'backward':
        return (f(a) - f(a - h)) / h
    elif method == 'central':
        return (f(a + h) - f(a - h)) / (2 * h)
```

```python
print("Derivative of e^x at x=0.5:", derivative(np.exp, 0.5))
print("Derivative of cos(x) at x=0.5:", derivative(np.cos, 0.5))
```
